```python
import numpy as np
import matplotlib.pyplot as plt
import sys

path = "saved_model/character_trajectories/char_l4_p2_interpretable/"
wpath = path + "weight-matrix.npy"
ppath= path + "prototypes.npy"

data = np.load(wpath, allow_pickle=True)
pdata = np.load(ppath, allow_pickle=True)
```

```python
def plotMatrix(mat):
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_xlabel('Prototypes')
    ax.set_ylabel('Classes')
    ax.matshow(mat, cmap='cool')
    for (i, j), z in np.ndenumerate(mat):
        ax.text(j, i, '{:0.1f}'.format(z),
                ha='center', va='center')
    plt.show()
    
def plotPatch(mat):
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_xlabel('Patch')
    ax.set_ylabel('Classes')
    ax.matshow(mat, cmap='cool')
    for (i, j), z in np.ndenumerate(mat):
        ax.text(j, i, '{:0.1f}'.format(z),
                ha='center', va='center')
    plt.show()
    
def plotProto(mat):
    mat = np.transpose(mat)
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_xlabel('Proto')
    ax.set_ylabel('Patch')
    ax.matshow(mat, cmap='cool')
    for (i, j), z in np.ndenumerate(mat):
        ax.text(j, i, '{:0.1f}'.format(z),
                ha='center', va='center')
    plt.show()
```

Maximum weight for each prototype for each class. Weight can differ based on the temporal position within the data. E.g. a prototype can be more important if it occurs at a specific location.

```python
plotMatrix(np.max(data, axis=-1))
```

Minimum weight for the prototypes. Shows that there is a difference in the importance based on the temporal position.

```python
plotMatrix(np.min(data, axis=-1))
```

Standard deviation for the prototype weights.

```python
plotMatrix(np.std(data, axis=-1))
```

Maximum difference between the prototype weights based on their temporal position.

```python
plotMatrix(np.max(data, axis=-1) - np.min(data, axis=-1))
```

Weights for the prototype 39.

```python
plotPatch(data[:,39,:])
```

Weights for class 18.

```python
plotProto(data[18,:,:])
```
